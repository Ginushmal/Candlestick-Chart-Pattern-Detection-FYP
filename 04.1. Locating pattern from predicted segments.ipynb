```python
import pandas as pd
import numpy as np


predicted_patterns = pd.read_csv('Datasets\predicted_patterns_inCont_tets_df.csv')
predicted_patterns
```

```python
# covert the index to the Segment_Id column in predicted_patterns_copy
predicted_patterns['Segment_Id'] = predicted_patterns.index
predicted_patterns
```

```python
origin_date = predicted_patterns['Pattern_Start'].astype('datetime64[ns]').min()
origin_date
```

```python
# take the smallest Pattern_Start date as the start date with value 0 , give a start and end date for each pattern as the distance from the start date in a new column
predicted_patterns['Pattern_Start_pos'] = predicted_patterns['Pattern_Start'].astype('datetime64[ns]').sub(predicted_patterns['Pattern_Start'].astype('datetime64[ns]').min()).dt.days
predicted_patterns['Pattern_End_pos'] = predicted_patterns['Pattern_End'].astype('datetime64[ns]').sub(predicted_patterns['Pattern_Start'].astype('datetime64[ns]').min()).dt.days
predicted_patterns
```

```python
# create a Centers column with the average of the Pattern_Start_pos and Pattern_End_pos
predicted_patterns['Center'] = (predicted_patterns['Pattern_Start_pos'] + predicted_patterns['Pattern_End_pos'])/2
predicted_patterns
```

```python
import pandas as pd
import numpy as np
from sklearn.cluster import DBSCAN


df = predicted_patterns.copy()

# Initialize a list to store merged clusters from all groups
merged_clusters = []

# Group by 'Chart Pattern' and apply clustering to each group
for pattern, group in df.groupby('Chart Pattern'):
    # print (pattern)
    # print(group)
    # Clustering
    centers = group['Center'].values.reshape(-1, 1)
    db = DBSCAN(eps=20, min_samples=10).fit(centers)
    group['Cluster'] = db.labels_
    
    # Filter out noise (-1) and group by Cluster
    for cluster_id, cluster_group in group[group['Cluster'] != -1].groupby('Cluster'):
        pattern_start = cluster_group['Pattern_Start_pos'].min()
        pattern_end = cluster_group['Pattern_End_pos'].max()
        merged_clusters.append({
            'Pattern_Start': origin_date + pd.Timedelta(days=pattern_start),
            'Pattern_End': origin_date + pd.Timedelta(days=pattern_end),
            'Chart Pattern': pattern,
            'Cluster': cluster_id,
            'Merged_Start': pattern_start,
            'Merged_End': pattern_end
        })

# Combine all merged clusters into a final DataFrame
merged_df = pd.DataFrame(merged_clusters)

# Display the result
merged_df
```
